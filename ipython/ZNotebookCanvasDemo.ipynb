This IPython Notebook demonstrates how graphics produced by ROOT **TObject** classes can be embedded  
inline within a IPython notebook. This capability is provided by my custom **ZNotebookCanvas** class,  
which is a part of the RIPN (ROOT IPython Notebook) module. The main difference is the user must call  
the ZNotebookCanvas::Insert() method *on the last line of a cell* to insert the canvas graphics inline.  

Note that the directory containing RIPN.py must be included in the Python path (i.e. the env. var. PYTHONPATH)

```python
from ROOT import ROOT, TH1F, TF1, TLegend, TLatex
from RIPN import ZNotebookCanvas
```

```python
# Create an instance of the special ZNotebookCanvas class, specifying
# the (x,y) size of the canvas/final image in the constructor
NC = ZNotebookCanvas()

# Create the histogram
H = TH1F('H','H',20,-4,4)
H.SetTitle("A gaussian histogram")
H.FillRandom('gaus',200)
H.SetStats(False)
H.GetXaxis().SetTitle("The X-axis!")
H.GetXaxis().SetTitleSize(0.06)
H.GetXaxis().CenterTitle()
H.GetXaxis().SetLabelSize(0.06)

H.GetYaxis().SetTitle("Counts")
H.GetYaxis().SetTitleSize(0.06)
H.GetYaxis().CenterTitle()
H.GetYaxis().SetLabelSize(0.06)

H.SetLineWidth(2)
H.SetMarkerStyle(20)
H.SetMarkerColor(ROOT.kBlue+3)
H.Draw("E0")

# Create the TF1 fit object and hit the histogram
F = TF1("fit","gaus",-3,3)
F.SetLineColor(ROOT.kRed)
F.SetLineWidth(2)
H.Fit("fit","R N")
F.Draw("C SAME")

# Create the Legend
L = TLegend(0.65, 0.65, 0.92, 0.92)
L.AddEntry(H,"Data","EPL")
L.AddEntry(F,"Fit","L")
L.Draw()

# Create the TLatex label to display the sigma value
FitParams = F.GetParameters()
sigma = FitParams[2]
Text = "#sigma = " + str(sigma)
T = TLatex(-3.5, 25, Text)
T.SetTextColor(ROOT.kRed)
T.Draw()

# Call the ZNotebookCanvas::Insert() method on the last line
# of the cell to insert the canvas graphics inline!
NC.Insert()
```

```python
# Let's do another example that demonstrates plotting in a divided
# ZNotebookCanvas. We'll make a 3x2 matrix of random gaussian
# histograms and draw them in different colors

# Create an array of colors
Colors = [ROOT.kRed, ROOT.kBlue+8, ROOT.kMagenta, 
          ROOT.kGreen+2, ROOT.kOrange, ROOT.kCyan]

# Create a ZNotebookCanvas object with a custom canvas
# size and then divide it (3 rows in x, 2 rows in y)
NC2 = ZNotebookCanvas(1200,400)
NC2.Divide(3,2)


# Now we'll create a TH1F and TF1 for the spectrum and fit,
# respectively, and iterate six times to complete the plot
H=[]
F=[]
num_hist = 6

for i in range(0,num_hist):
    
    # Switch to the appropriate subcanvas
    NC2.cd(i+1)
    
    H.append(TH1F('H','H',30,-5,5))
    H[i].FillRandom('gaus',100)
    H[i].GetXaxis().SetLabelSize(0.1)
    H[i].GetYaxis().SetLabelSize(0.1)
    H[i].SetFillColor(Colors[i])
    H[i].SetStats(False)    
    H[i].Draw('B')

    F.append(TF1('Fit','gaus',-4,4))
    F[i].SetLineColor(1)
    F[i].SetLineWidth(2)
    F[i].SetLineStyle(1)
      
    H[i].Fit('Fit','R N')
    F[i].Draw('SAME C')
    
# Insert the finished graphic into the  notebook
NC2.Insert()
```
